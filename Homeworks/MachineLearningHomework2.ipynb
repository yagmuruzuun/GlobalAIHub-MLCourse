<h3> Checking outliers with Isolation Forest </h3>
Due to the high number of outliers in IsolationForest, we can assume that this values are not outliers. We can go without removing them to see performance.

```python
# Outlier detection
from sklearn.ensemble import IsolationForest
import numpy as np
temp = df.drop("output", axis=1)

outliers = IsolationForest(random_state=0).fit_predict(temp)
outliers_index = list(np.where(outliers == -1)[0])

del temp
print(f"Outlier Count: {len(outliers_index)} \nSample Count: {len(df)} \nFraction: {round(len(outliers_index)/len(df),3)}")
```

<h3> Creating input/output </h3>

```python
X=df.drop(["output","s2"],axis=1)
y=df[["output"]]
```

```python
X.head(3)
```

```python
y.head(3)
```

<h3>Dividing dataset into train and test </h3>
Splitting as 0.7 train - 0.3 test ratio

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

<h2> MODEL SELECTION </h2>

<h3> Linear Regression </h3>

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(X_train, y_train)
y_pred_reg=linreg.predict(X_test)
```

```python
coeff = pd.DataFrame(linreg.coef_,columns = X.columns)
print("Intercept is:",linreg.intercept_,"\nFeature coefficients are:")
coeff
```

```python
from sklearn.metrics import mean_squared_error
reg_train=linreg.score(X_train, y_train)
reg_test=linreg.score(X_test, y_test)
reg_mse=mean_squared_error(y_pred_reg,y_test)
print("Linear Regression Train Score: {} \nLinear Regression Test Score: {} \nMean Squared Error: {}".format(reg_train,reg_test,reg_mse))
```

<h3> Ridge Regression </h3>

```python
from sklearn.linear_model import Ridge

ridge_model1 = Ridge(alpha = 0.05)
ridge_model1.fit(X_train, y_train)
y_pred_ridge1 = ridge_model1.predict(X_test)
print(f'Ridge model 1 coef: {ridge_model1.coef_}')
ridge_model2 = Ridge(alpha = 0.1) 
ridge_model2.fit(X_train, y_train)
y_pred_ridge2 = ridge_model2.predict(X_test)
print(f'Ridge model 2 coef: {ridge_model2.coef_}')
ridge_model3 = Ridge(alpha = 0.2)
ridge_model3.fit(X_train, y_train)
y_pred_ridge3 = ridge_model3.predict(X_test)
print(f'Ridge model 3 coef: {ridge_model3.coef_}')
ridge_model4 = Ridge(alpha = 0.3)
ridge_model4.fit(X_train, y_train)
y_pred_ridge4 = ridge_model4.predict(X_test)
print(f'Ridge model 4 coef: {ridge_model4.coef_}')
ridge_model5 = Ridge(alpha = 0.4)
ridge_model5.fit(X_train, y_train)
y_pred_ridge5 = ridge_model5.predict(X_test)
print(f'Ridge model 5 coef: {ridge_model5.coef_}')
```

```python
#Score
ridge_1_train, ridge_1_test, ridge_1_mse = ridge_model1.score(X_train, y_train), ridge_model1.score(X_test, y_test), mean_squared_error(y_pred_ridge1,y_test)
ridge_2_train, ridge_2_test, ridge_2_mse = ridge_model2.score(X_train, y_train), ridge_model2.score(X_test, y_test), mean_squared_error(y_pred_ridge2,y_test)
ridge_3_train, ridge_3_test, ridge_3_mse = ridge_model3.score(X_train, y_train), ridge_model3.score(X_test, y_test), mean_squared_error(y_pred_ridge3,y_test)
ridge_4_train, ridge_4_test, ridge_4_mse = ridge_model4.score(X_train, y_train), ridge_model4.score(X_test, y_test), mean_squared_error(y_pred_ridge4,y_test)
ridge_5_train, ridge_5_test, ridge_5_mse = ridge_model5.score(X_train, y_train), ridge_model5.score(X_test, y_test), mean_squared_error(y_pred_ridge5,y_test)
```

<h3> Lasso Regression </h3>

```python
from sklearn.linear_model import Lasso

lasso_model1 = Lasso(alpha = 0.05)
lasso_model1.fit(X_train, y_train)
y_pred_lasso1 = lasso_model1.predict(X_test)
print(f'Lasso model 1 coef: {lasso_model1.coef_}')
lasso_model2 = Lasso(alpha = 0.1)
lasso_model2.fit(X_train, y_train)
y_pred_lasso2 = lasso_model2.predict(X_test)
print(f'Lasso model 2 coef: {lasso_model2.coef_}')
lasso_model3 = Lasso(alpha = 0.2)
lasso_model3.fit(X_train, y_train)
y_pred_lasso3 = lasso_model3.predict(X_test)
print(f'Lasso model 3 coef: {lasso_model3.coef_}')
lasso_model4 = Lasso(alpha = 0.3) 
lasso_model4.fit(X_train, y_train)
y_pred_lasso4 = lasso_model4.predict(X_test)
print(f'Lasso model 4 coef: {lasso_model4.coef_}')
lasso_model5 = Lasso(alpha = 0.4)
lasso_model5.fit(X_train, y_train)
y_pred_lasso5 = lasso_model5.predict(X_test)
print(f'Lasso model 5 coef: {lasso_model5.coef_}')
```

```python
#Score
lasso_1_train, lasso_1_test, lasso_1_mse = lasso_model1.score(X_train, y_train), lasso_model1.score(X_test, y_test), mean_squared_error(y_pred_lasso1,y_test)
lasso_2_train, lasso_2_test, lasso_2_mse = lasso_model2.score(X_train, y_train), lasso_model2.score(X_test, y_test), mean_squared_error(y_pred_lasso2,y_test)
lasso_3_train, lasso_3_test, lasso_3_mse = lasso_model3.score(X_train, y_train), lasso_model3.score(X_test, y_test), mean_squared_error(y_pred_lasso3,y_test)
lasso_4_train, lasso_4_test, lasso_4_mse = lasso_model4.score(X_train, y_train), lasso_model4.score(X_test, y_test), mean_squared_error(y_pred_lasso4,y_test)
lasso_5_train, lasso_5_test, lasso_5_mse = lasso_model5.score(X_train, y_train), lasso_model5.score(X_test, y_test), mean_squared_error(y_pred_lasso5,y_test)
```

<h2> RESULT </h2>

```python
train = [reg_train, ridge_1_train, ridge_2_train, ridge_3_train, ridge_4_train, ridge_5_train, lasso_1_train, lasso_2_train, lasso_3_train, lasso_4_train, lasso_5_train]
test = [reg_test, ridge_1_test, ridge_2_test, ridge_3_test, ridge_4_test, ridge_5_test, lasso_1_test, lasso_2_test, lasso_3_test, lasso_4_test, lasso_5_test]
mse = [reg_mse, ridge_1_mse, ridge_2_mse, ridge_3_mse, ridge_4_mse, ridge_5_mse, lasso_1_mse, lasso_2_mse, lasso_3_mse, lasso_4_mse, lasso_5_mse]
model = ["Linear Regression","Ridge Regression(alpha=0.05)","Ridge Regression(alpha=0.1)","Ridge Regression(alpha=0.2)","Ridge Regression(alpha=0.3)","Ridge Regression(alpha=0.4)","Lasso Regression(alpha=0.05)","Lasso Regression(alpha=0.1)","Lasso Regression(alpha=0.2)","Lasso Regression(alpha=0.3)","Lasso Regression(alpha=0.4)"]
```

```python
scores = pd.DataFrame({"Model Name":model,"Train Score":train,"Test Score":test,"MSE":mse})
scores
```
